```python
#function for avoiding repetition of code 
def display_plot(df,col,col1,title):
    x = sns.barplot(x = df[col] ,y = df[col1],data =df , palette= 'rocket')
    x.set_title(title)
    plt.xticks(rotation = 90)
    return plt.show()
```

```python
#function for avoiding repetition of code 
def countplay_plot(df,col,title):
    x = sns.countplot(x = df[col] ,hue = df.sex,data =df , palette= 'viridis')
    x.set_title(title)
    plt.xticks(rotation = 90)
    return plt.show()
```

```python
def line_plot(df,col,col1,title):
    plt.figure(figsize=(15,5))
    x= sns.lineplot(x=df[col],y=df[col1],hue=df.sex,data=df)
    x.set_title(title)
    plt.show()
```

# Data Preprocessing

# Steps 

- Check the columns 

- feature engneering

- Spliting Data to three sets due to year column 

- Delete the useless columns by make now data frames for sets 

- Standard Scaler for numeric columns 

- one hot encoder for categorical columns 

- merge two of them 

```python
df.head()
```

```python
cols = ['country', 'sex', 'age', 'generation']
for col in cols:
    print(df[col].unique())
    print('-' * 50)
```

```python
mean = df['suicides/100k pop'].mean()
df['danger'] = df['suicides/100k pop'].apply(lambda x : 'High' if x > mean else 'Low')
```

- appling feature engneering on suicides/100k pop to use classification later on the new columns called danger has two values High and Low according to mean

```python
year_df = df['year'].value_counts()
year_df.sort_index(inplace = True)
year_cum_precentage = year_df.cumsum() / year_df.sum() * 100
year_cum_precentage
```

- according to year columns with cumulative sum we notice the size of data, so we are gonna split the dataset into three sets train, validation, and test

```python
train_df = df.loc[df['year'] <= 2006]
validation_df = df.loc[(df['year'] >= 2007) & (df['year'] <=2011)]
test_df = df.loc[df['year'] >= 2012]

print('train_df',train_df.shape)
print('validation_df',validation_df.shape)
print('test_df',test_df.shape)
print('train_df:',train_df.shape[0] / df.shape[0] * 100)
print('validation_df:',validation_df.shape[0] / df.shape[0] * 100)
print('test_df:',test_df.shape[0] / df.shape[0] * 100)
```

- we are gonna make inputs and target dataframes and put the columns that we will use instead of drop the others 
- Year : It will be not relevant as prediction in future 
- sucicides_no : have redundant information with suicides/100k pop
- suicides/100k pop : have redundant information with the new columns danger 
- generation : have redundant information with age 

```python
input_cols = ['country', 'sex', 'age','population', 'HDI for year', 'gdp_per_capita']
target_col = 'danger'
```

```python
x_train = train_df.loc[:, input_cols]
y_train = train_df.loc[:, target_col]

x_validation = validation_df.loc[:, input_cols]
y_validation = validation_df.loc[:, target_col]

x_test = test_df.loc[:, input_cols]
y_test = test_df.loc[:, target_col]
```

```python
numeric_cols = list(x_train.select_dtypes(include=np.number).columns)
categorical_cols = list(x_train.select_dtypes(include='object').columns)

print('numeric columns: ', numeric_cols)
print('categorical columns: ', categorical_cols)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df[numeric_cols])

x_train.loc[:, numeric_cols] = scaler.transform(x_train.loc[:, numeric_cols])
x_validation.loc[:, numeric_cols] = scaler.transform(x_validation.loc[:, numeric_cols])
x_test.loc[:, numeric_cols] = scaler.transform(x_test.loc[:, numeric_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoder.fit(df[categorical_cols])
```

```python
encoded_cols = list(encoder.get_feature_names(categorical_cols))

x_train[encoded_cols] = encoder.transform(x_train.loc[:, categorical_cols]).toarray()
x_validation[encoded_cols] = encoder.transform(x_validation.loc[:, categorical_cols]).toarray()
x_test[encoded_cols] = encoder.transform(x_test.loc[:, categorical_cols]).toarray()
```

- Now we are gonna concatenate two types numeric and categorical

```python
x_train = x_train[numeric_cols + encoded_cols]
x_validation = x_validation[numeric_cols + encoded_cols]
x_test = x_test[numeric_cols + encoded_cols]
```

# Model Buliding

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score , f1_score , classification_report , confusion_matrix
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
train_pred = model.predict(x_train)
print('trainset accuracy:',accuracy_score(y_train, train_pred))
print('classification report: \n',classification_report(y_train , train_pred))
print('trainset Confusion Matrix: \n',confusion_matrix(y_train, train_pred))
```

```python
validation_pred = model.predict(x_validation)
accuracy_score(y_validation, validation_pred)
print('validationset accuracy:',accuracy_score(y_validation, validation_pred))
print('classification report: \n',classification_report(y_validation, validation_pred))
print('validationset Confusion Matrix: \n',confusion_matrix(y_validation, validation_pred))
```

```python
x = pd.concat([x_train, x_validation])
y = pd.concat([y_train, y_validation])
print(x.shape, y.shape)
```

```python
from sklearn.model_selection import GridSearchCV
param_dict = {
    'criterion': ['gini', 'entropy'],
    "max_depth": range(2, 40, 2),
    "min_samples_split": range(2, 10, 2)
}
model = DecisionTreeClassifier()
grid = GridSearchCV(estimator = model, param_grid=param_dict,scoring='accuracy',refit=True,cv= 6,verbose= 1,n_jobs=-1)
grid.fit(x, y)
```

```python
print('best params: ', grid.best_params_)
print('accuracy score: ', grid.best_score_)
```

```python
test_pred = grid.predict(x_test)
accuracy_score(y_test, test_pred)
```

```python
def predict_input(model, single_input):
    input_df = pd.DataFrame([single_input])
    input_df[model['numeric_cols']] = model['scaler'].transform(input_df[model['numeric_cols']])
    input_df[model['encoded_cols']] = model['encoder'].transform(input_df[model['categorical_cols']]).toarray()
    X_input = input_df[model['numeric_cols'] + model['encoded_cols']]
    pred = model['model'].predict(X_input)[0]
    prob = model['model'].predict_proba(X_input)[0][list(model['model'].classes_).index(pred)]
    return pred, prob
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(8,4)})
```

```python
df = pd.read_csv('master.csv')
print(df.shape)
df.head()
```

```python
df[df['HDI for year'] ==  NaN]
```

```python
data_country = pd.DataFrame(df['country'].value_counts(normalize= False)).reset_index()
pd.options.display.max_rows = None
data_country
```

```python
lista = data_country[data_country['country'] >=  50]['index'].tolist()
```

```python
def country_threshold(x):
    if x in lista:
        return x
    else:
        return 'other'
df['new countries'] = df['country'].apply(country_threshold)    
```

```python
df.head()
```

```python
new_df = df.drop(df[df['new countries'] == 'other'].index ,axis = 0)
print(new_df.shape)
```

```python
new_df['new countries'].value_counts()
```

```python
def danger_levels(x):
    if x < 5:
        return 0
    elif 9 > x > 5:
        return 1
    elif 14.9 > x > 10.0:
        return 2
    else:
        return 3
df['danger'] = df['suicides/100k pop'].apply(danger_levels)
```

```python
df.head()
```

```python
df['danger'].value_counts() / df.shape[0] * 100
```

```python
df.drop(columns = ['country','suicides_no','suicides/100k pop','country-year','HDI for year',' gdp_for_year'], axis = 1 , inplace = True)
```

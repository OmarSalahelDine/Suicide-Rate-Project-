```python
df['country'].unique()
```

```python
df.loc[:, 'country'].replace({'Czech Republic': 'Czechia', 
                              'Republic of Korea': 'Korea',
                              'Saint Vincent and Grenadines': 'Saint Vincent and the Grenadines'}, inplace = True)
df['country'].unique()
```

```python
for x in df['country'].unique():
#     print(x, end = ': ')
    if hdi_df.index.str.contains(x).any():
#         print(1)
        pass
    else:
#         print(0)
        print(f'{x}: 0')
```

```python
#drop Aruba
#drop Macao
#drop Puerto Rico
#drop San Marino
```

```python
filt = df['country'].apply(lambda x: x in ['Aruba', 'Macau', 'Puerto Rico', 'San Marino'])
indices = df.loc[filt].index
df.drop(indices, inplace = True)
df.shape
```

```python
indices = []
for x in df['country'].unique():
    exists = hdi_df.index.str.contains(x).any()
    if exists:
        row_index = hdi_df.index.str.contains(x).argmax()
        indices.append(row_index)
```

```python
hdi_df.index.str.contains('Egypt').argmax()
```

```python
hdi_df = hdi_df.iloc[indices]
hdi_df
```

```python
hdi_df.shape
```

```python
df.shape
```

```python
hdi_df.index = hdi_df.index.str.split('(').str[0].str.strip()
hdi_df
```

```python
for x in hdi_df.index:
    print(x, end = ': ')
    if not hdi_df.loc[x].isna().any():
        print(1)
    else:
        print(0)
```

```python
import matplotlib.pyplot as plt

for x in hdi_df.index:
    if hdi_df.loc[x].isna().any():
        row_series = hdi_df.loc[x]
        print(x)
        print('-' * 30)
        row_series.plot(kind = 'line')
        plt.show()
        print('=' * 60)
```

```python
filt = df['year'] < 1990
indices = df[filt].index
df.drop(indices, inplace = True)
```

```python
df['HDI for year'] = df.apply(lambda x: hdi_df.loc[x['country'], str(x['year'])] if pd.isna(x['HDI for year']) else x['HDI for year'], axis = 1)
df.head()
```

```python
df.dropna(inplace = True)
df.info()
```
